# Q1. displaying all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
! pip install bs4
!pip install requests
```

```python
import requests
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.status_code
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
Soup = BeautifulSoup(page.content,'html.parser')
```

```python
headings=Soup.find_all('span',class_="mw-headline")
headings[0:]
```

Q.7:. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The 
scraped data should include Product Name, Price, Image URL and Average Rating.

```python
import requests
page = requests.get("https://www.amazon.in/s?k=mobile+phones+under+20000&ref=nb_sb_noss_2")
page
```

```python
page.status_code
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
Soup = BeautifulSoup(page.content, 'html.parser')
```

```python
products = Soup.find_all('span', class_="a-size-medium a-color-base a-text-normal")
products[0:]
```

Finding the prices of the products

```python
Price = Soup.find_all('span', class_="a-price-whole")
Price[0:]
```

Finding ratings of the products

```python
Ratings = Soup.find_all('span', class_="a-icon-alt")
Ratings[0:]
```

Images URL

```python
Images = Soup.find_all('span', class_="https://m.media-amazon.com/images/I/71zbz1O+hqL._AC_UY218_.jpg")
Images[0:]
```
